```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sd_prop_crime_path = "san_diego/san_diego_all_property_crimes_2000_2021.csv"
sd_violent_crime_path = "san_diego/san_diego_all_violent_crimes_2000_2021.csv"
sd_burglary_crime_path = "san_diego/san_diego_burglary_2000_2021.csv"
la_prop_crime_path = "los_angeles/la_all_property_crimes_2000_2021.csv"
la_violent_crime_path = "los_angeles/la_all_violent_crime_2000_2021.csv"
la_burglary_crime_path = "los_angeles/la_burglary_2000_2021.csv"
slt_prop_crime_path = "south_lake_tahoe/slt_all_property_crime_2000_2021.csv"
slt_violent_crime_path = "south_lake_tahoe/slt_all_violent_crime_2000_2021.csv"
slt_burglary_crime_path = "south_lake_tahoe/slt_all_burglary_2000_2021.csv"
```

```python
sd_prop_crime = pd.read_csv(sd_prop_crime_path)
sd_violent_crime = pd.read_csv(sd_violent_crime_path)
sd_burglary_crime = pd.read_csv(sd_burglary_crime_path)
la_prop_crime = pd.read_csv(la_prop_crime_path)
la_violent_crime = pd.read_csv(la_violent_crime_path)
la_burglary_crime = pd.read_csv(la_burglary_crime_path)
slt_prop_crime = pd.read_csv(slt_prop_crime_path)
slt_violent_crime = pd.read_csv(slt_violent_crime_path)
slt_burglary_crime = pd.read_csv(slt_burglary_crime_path)
```

```python
sd1_df = pd.DataFrame(sd_prop_crime)
sd2_df = pd.DataFrame(sd_violent_crime)
sd3_df = pd.DataFrame(sd_burglary_crime)
la1_df = pd.DataFrame(la_prop_crime)
la2_df = pd.DataFrame(la_violent_crime)
la3_df = pd.DataFrame(la_burglary_crime)
slt1_df = pd.DataFrame(slt_prop_crime)
slt2_df = pd.DataFrame(slt_violent_crime)
slt3_df = pd.DataFrame(slt_burglary_crime)
```

```python
sd1_df.insert(0,"Police Department",["San Diego", "San Diego"])
sd2_df.insert(0,"Police Department",["San Diego", "San Diego"])
sd3_df.insert(0,"Police Department",["San Diego", "San Diego"])
sd1_df.insert(1,"Type of Crime",["All Property", "All Property"])
sd2_df.insert(1,"Type of Crime",["All Violent", "All Violent"])
sd3_df.insert(1,"Type of Crime",["Burglary", "Burglary"])
```

```python
la1_df.insert(0,"Police Department",["Los Angeles", "Los Angeles"])
la2_df.insert(0,"Police Department",["Los Angeles", "Los Angeles"])
la3_df.insert(0,"Police Department",["Los Angeles", "Los Angeles"])
la1_df.insert(1,"Type of Crime",["All Property", "All Property"])
la2_df.insert(1,"Type of Crime",["All Violent", "All Violent"])
la3_df.insert(1,"Type of Crime",["Burglary", "Burglary"])
```

```python
slt1_df.insert(0,"Police Department",["South Lake Tahoe", "South Lake Tahoe"])
slt2_df.insert(0,"Police Department",["South Lake Tahoe", "South Lake Tahoe"])
slt3_df.insert(0,"Police Department",["South Lake Tahoe", "South Lake Tahoe"])
slt1_df.insert(1,"Type of Crime",["All Property", "All Property"])
slt2_df.insert(1,"Type of Crime",["All Violent", "All Violent"])
slt3_df.insert(1,"Type of Crime",["Burglary", "Burglary"])
```

```python
sd_complete = pd.concat([sd1_df, sd2_df, sd3_df])
la_complete = pd.concat([la1_df, la2_df, la3_df])
slt_complete = pd.concat([slt1_df, slt2_df, slt3_df])
```

```python
sd_complete = sd_complete.reset_index(drop=True)
la_complete = la_complete.reset_index(drop=True)
slt_complete = slt_complete.reset_index(drop=True)
```

```python
sd_complete
```

```python
sd_complete_reported = sd_complete.loc[sd_complete["series"]=="Reported"]
la_complete_reported = la_complete.loc[la_complete["series"]=="Reported"]
slt_complete_reported = slt_complete.loc[slt_complete["series"]=="Reported"]
x_axis = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
sd_total_crime = []
la_total_crime = []
slt_total_crime = []
```

```python
for x in x_axis : 
    x=str(x)
    sum_la = la_complete_reported [x].sum()
    la_total_crime.append(sum_la)
```

```python
for x in x_axis : 
    x=str(x)
    sum_sd = sd_complete_reported [x].sum()
    sd_total_crime.append(sum_sd)
```

```python
for x in x_axis : 
    x=str(x)
    sum_slt = slt_complete_reported [x].sum()
    slt_total_crime.append(sum_slt)
```

```python
sd_total_crime_line = plt.plot(x_axis, sd_total_crime, color = "red", label = "San Diego Total Crime")
plt.title("Reported Crimes YoY San Diego")
plt.xlabel("Year")
plt.ylabel("Reported Crime Count")
plt.grid()
```

```python
la_total_crime_line = plt.plot(x_axis, la_total_crime, color = "blue", label = "Los Angeles Total Crime")
plt.title("Reported Crimes YoY Los Angeles")
plt.xlabel("Year")
plt.ylabel("Reported Crime Count")
plt.grid()
```

```python
sd_total_crime_line = plt.plot(x_axis, slt_total_crime, color = "green", label = "South Lake Tahoe Total Crime")
plt.title("Reported Crimes YoY South Lake Tahoe")
plt.xlabel("Year")
plt.ylabel("Reported Crime Count")
plt.grid()
```

```python
sd_total_crime_line = plt.plot(x_axis, sd_total_crime, color = "red", label = "San Diego Total Crime")
sd_total_crime_line = plt.plot(x_axis, la_total_crime, color = "blue", label = "Los Angeles Total Crime")
sd_total_crime_line = plt.plot(x_axis, slt_total_crime, color = "green", label = "South Lake Tahoe Total Crime")
plt.title("Reported Crimes YoY All 3 Locations")
plt.xlabel("Year")
plt.ylabel("Reported Crime Count")
plt.grid()
```
