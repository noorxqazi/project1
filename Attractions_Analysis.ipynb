# Attractions Near Each Home
---

### Analysis

##### After pulling data for the best 10 attractions in each city within 20 miles of each property, we can determine the following:
* Our Los Angeles home has the closest attractions nearby, with 9/10 being within 10 miles of the property. 
* Attractions in San Diego are the furthest away from our property with the closest one being 7.8 miles from the home. 
* South Lake Tahoe is the only residence that has a 5-Star attraction within 10 miles of the home. 
* South Lake Tahoe and San Diego have the greatest variety of attractions nearby, where as 50% of the attractions nearby our LA home are museums.

```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Store filepath in a variable
attractions = "Resources/attractions_data.csv"

attractions_df = pd.read_csv(attractions, encoding="ISO-8859-1")
```

```python
attractions_df
```

```python
# Show only data for LA homes 
LA_home = attractions_df.loc[attractions_df["City"] == 'Los Angeles']
LA_home.head()
```

```python
# Sort LA attractions by distance to home 

closest_LA = LA_home.sort_values('Distance to Home (miles)', ascending=True)
closest_LA
```

```python
# Count unique categories for Los Angeles attractions
LA_categories = closest_LA["Category"].value_counts()
LA_categories
```

```python
# Create a pie graph for each category. 
LA_categories.plot(kind="pie", autopct = "%1.1f%%")
plt.title = "Los Angeles Attraction Categories"
plt.savefig("Resources/LA_Attraction_Categories.png")
plt.show()
```

```python
# Show only data for South Lake Tahoe home
SLT_home = attractions_df.loc[attractions_df["City"] == 'South Lake Tahoe']
SLT_home.head()
```

```python
# Sort South Lake Tahoe attractions by distance to home 

closest_SLT = SLT_home.sort_values('Distance to Home (miles)', ascending=True)
closest_SLT
```

```python
# Count unique categories for South Lake Tahoe attractions
SLT_categories = closest_SLT["Category"].value_counts()
SLT_categories
```

```python
SLT_categories.plot(kind="pie", autopct = "%1.1f%%")
plt.savefig("Resources/SLT_Attraction_Categories.png")
plt.show()
```

```python
# Show only data for San Diego home
SD_home = attractions_df.loc[attractions_df["City"] == 'San Diego']
SD_home.head()
```

```python
# Sort San Diego attractions by distance to home 

closest_SD = SD_home.sort_values('Distance to Home (miles)', ascending=True)
closest_SD
```
